# bronze airports

```python
airports_bronze = spark.table("bronze_airports")

display(airports_bronze.limit(5))
```

```python
from pyspark.sql.functions import col

silver_airports = (
    airports_bronze
    .select(
        col("_c0").alias("airport_id"),
        col("_c1").alias("airport_name"),
        col("_c2").alias("city"),
        col("_c3").alias("country"),
        col("_c4").alias("iata_code"),
        col("_c5").alias("icao_code"),
        col("_c6").alias("latitude"),
        col("_c7").alias("longitude")
    )
)
```

```python
silver_airports = (
    silver_airports
    .filter(col("iata_code") != "\\N")
)
```

```python
silver_airports.count()
```

```python
silver_airports.write \
.mode("overwrite") \
.saveAsTable("silver_airports")
```

```python
bts = spark.table("bronze_bts_flights")
```

```python
airport_join_test = (
    bts.alias("b")
    .join(
        silver_airports.alias("a"),
        col("b.ORIGIN") == col("a.iata_code"),
        "inner"
    )
)
```

```python
airport_join_test.count()
```

```python
display(
    airport_join_test.select(
        "ORIGIN",
        "airport_name",
        "city"
    )
)
```

```python
from pyspark.sql.functions import to_timestamp,to_date,col

silver_flights = (
    bts
    .withColumn(
        "flight_timestamp",
        to_timestamp(
            col("FL_DATE"),
            "M/d/yyyy hh:mm:ss a"
        )
    )
    .withColumn(
        "flight_date",
        to_date("flight_timestamp")
    )
)
```

```python
display(
    silver_flights.select(
        "FL_DATE",
        "flight_timestamp",
        "flight_date"
    ).limit(20)
)
```

```python
silver_flights.filter(
    col("flight_date").isNull()
).count()
```

```python
silver_flights = (
    silver_flights.select(
        "flight_date",
        "OP_UNIQUE_CARRIER",
        "ORIGIN",
        "DEST",
        "DEP_DELAY",
        "ARR_DELAY",
        "CANCELLED",
        "DIVERTED",
        "AIR_TIME",
        "DISTANCE"
    )
)
```

```python
silver_flights.write \
.mode("overwrite") \
.saveAsTable("silver_flights")
```

# AIRLINE CONFORMANCE DIAGNOSTIC

```python
from pyspark.sql.functions import col

bts = spark.table("bronze_bts_flights")

print("Bronze BTS Rows:", bts.count())
bts.printSchema()
```

```python
airline_columns = [
    c for c in bts.columns
    if "AIRLINE" in c.upper() or "CARRIER" in c.upper()
]

airline_columns
```

```python
bts.select(
    *[col(c) for c in airline_columns]
).distinct().show(100, truncate=False)
```
