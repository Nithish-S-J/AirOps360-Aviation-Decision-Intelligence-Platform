# AirOps 360

## Bronze Layer Ingestion

Source:
- OpenFlights

Files:
- airports.dat
- airlines.dat
- routes.dat

Layer:
- Bronze

Purpose:
- Raw Master Data Ingestion

```python
from pyspark.sql import functions as F
```

# Airports

```python
airports_df = (
    spark.read
    .option("header","false")
    .option("inferSchema","true")
    .csv("Files/raw/reference/openflights/airports.dat")
)
```

```python
airports_df.count()
```

```python
display(airports_df.limit(10))
```

```python
airports_df.printSchema()
```

```python
bronze_airports = (
    airports_df
    .withColumn(
        "ingestion_timestamp",
        F.current_timestamp()
    )
    .withColumn(
        "source_system",
        F.lit("OPENFLIGHTS")
    )
)
```

```python
bronze_airports.write \
.mode("overwrite") \
.saveAsTable("bronze_airports")
```

```python
display(
    spark.table("bronze_airports")
)
```
