# Import Libraries

```python
from pyspark.sql import functions as F
from pyspark.sql.types import *
```

```python
bts_df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("Files/raw/batch/bts/T_ONTIME_REPORTING.csv")
)

```

```python
print("Row Count:", bts_df.count())
```

```python
display(bts_df.limit(10))
```

# BTS Profiling

```python
bts_df.count()
```

```python
bts_df.select("OP_UNIQUE_CARRIER").distinct().count()
```

```python
bts_df.filter(F.col("DIVERTED") == 1).count()
```

```python
bronze_bts = (
    bts_df
    .withColumn("ingestion_timestamp", F.current_timestamp())
    .withColumn("source_system", F.lit("BTS"))
)
```

```python
display(bronze_bts.limit(5))
```

```python
bronze_bts.write.mode("overwrite").saveAsTable("bronze_bts_flights")
```
