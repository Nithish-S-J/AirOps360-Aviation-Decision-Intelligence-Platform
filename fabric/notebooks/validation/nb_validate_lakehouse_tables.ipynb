```python
from pyspark.sql import functions as F

tables = [
    t.name
    for t in spark.catalog.listTables()
    if not t.isTemporary
]

print("TOTAL LAKEHOUSE TABLES:", len(tables))
print("=" * 100)

for table_name in sorted(tables):
    try:
        df = spark.table(table_name)

        print(f"\nTABLE: {table_name}")
        print(f"ROWS: {df.count():,}")
        print(f"COLUMNS: {len(df.columns)}")
        print("COLUMN LIST:")
        print(", ".join(df.columns))
        print("-" * 100)

    except Exception as e:
        print(f"\nTABLE: {table_name}")
        print(f"ERROR: {str(e)}")
        print("-" * 100)
```

```python
silver_capacity = spark.table("silver_faa_airport_capacity")
fact_capacity = spark.table("fact_airport_capacity")

silver_minus_fact = silver_capacity.exceptAll(fact_capacity).count()
fact_minus_silver = fact_capacity.exceptAll(silver_capacity).count()

print("Silver Minus Fact Rows:", silver_minus_fact)
print("Fact Minus Silver Rows:", fact_minus_silver)
```

```python
tables_to_drop = [
    "gold_airline_operations",
    "gold_airport_operations",
    "gold_network_summary",
    "gold_route_operations",
    "silver_faa_airport_capacity"
]

for table_name in tables_to_drop:
    spark.sql(f"DROP TABLE IF EXISTS {table_name}")
    print(f"Dropped: {table_name}")

print("Lakehouse cleanup completed.")
```
